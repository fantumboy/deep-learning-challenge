## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
updated_df = application_df.drop(['EIN', 'NAME'], axis = 1)
updated_df.head()
```

```python
# Determine the number of unique values in each column.
unique_counts_column = updated_df.nunique()
unique_counts_column
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_value_counts = updated_df['APPLICATION_TYPE'].value_counts()
application_type_value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# application_types_to_replace = updated_df['APPLICATION_TYPE']

name_counts = updated_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(name_counts[name_counts <= 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    updated_df['APPLICATION_TYPE'] = updated_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
updated_df['APPLICATION_TYPE'].value_counts()


# name_counts = application_df['NAME'].value_counts()
# application_type_to_replace = list(name_counts[name_counts <= 500].index)


# # Replace in dataframe
# for app in application_type_to_replace:
#     application_df['NAME'] = application_df['NAME'].replace(app,"Other")


# Check to make sure binning was successful
# updated_df['NAME'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_value_counts = updated_df['CLASSIFICATION'].value_counts()
classification_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_filtered = classification_value_counts[classification_value_counts > 1]
class_filtered
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_counts = updated_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    updated_df['CLASSIFICATION'] = updated_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
updated_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
columns_to_convert = ['APPLICATION_TYPE',	'CLASSIFICATION', 'AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']
df_converted = pd.get_dummies(updated_df, columns = columns_to_convert)
df_converted
```

```python
# Split our preprocessed data into our features and target arrays
X = df_converted[['STATUS',
                  'ASK_AMT',
                  'APPLICATION_TYPE_Other',
                  'APPLICATION_TYPE_T10',
                  'APPLICATION_TYPE_T19',
                  'APPLICATION_TYPE_T3',
                  'APPLICATION_TYPE_T4',
                  'APPLICATION_TYPE_T5',
                  'APPLICATION_TYPE_T6',
                  'APPLICATION_TYPE_T7',
                  'APPLICATION_TYPE_T8',
                  'CLASSIFICATION_C1000',
                  'CLASSIFICATION_C1200',
                  'CLASSIFICATION_C2000',
                  'CLASSIFICATION_C2100',
                  'CLASSIFICATION_C3000',
                  'CLASSIFICATION_Other',
                  'AFFILIATION_CompanySponsored',
                  'AFFILIATION_Family/Parent',
                  'AFFILIATION_Independent',
                  'AFFILIATION_National',
                  'AFFILIATION_Other',
                  'AFFILIATION_Regional',
                  'USE_CASE_CommunityServ',
                  'USE_CASE_Heathcare',
                  'USE_CASE_Other',
                  'USE_CASE_Preservation',
                  'USE_CASE_ProductDev',
                  'ORGANIZATION_Association',
                  'ORGANIZATION_Co-operative',
                  'ORGANIZATION_Corporation',
                  'ORGANIZATION_Trust',
                  'INCOME_AMT_0',
                  'INCOME_AMT_1-9999',
                  'INCOME_AMT_10000-24999',
                  'INCOME_AMT_100000-499999',
                  'INCOME_AMT_10M-50M',
                  'INCOME_AMT_1M-5M',
                  'INCOME_AMT_25000-99999',
                  'INCOME_AMT_50M+',
                  'INCOME_AMT_5M-10M',
                  'SPECIAL_CONSIDERATIONS_N',
                  'SPECIAL_CONSIDERATIONS_Y']]

y = df_converted['IS_SUCCESSFUL']

# list(df_converted.columns)

# print(df_converted['INCOME_AMT'].value_counts())

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=3))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
train_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file

train_model.save('train_model.h5')

# loaded_model = load_model('your_model.h5')
# loaded_model.summary()
```
